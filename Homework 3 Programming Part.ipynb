```python
#Question 1d
import pylab
import numpy as np
import matplotlib.pyplot as plt
from numpy import cos
def myformat(x):
    return ('%.3g' % x)
def gd_hw2_2(x0,alpha,iter):
    x1=x0
    print ("When alpha=",alpha,"x0=",x0)
    for i in range(iter):
        k=x1[0]-alpha*(6*x1[0]+4*x1[1]-5)
        x1[1]=x1[1]-alpha*(4*x1[0]+4*x1[1])
        x1[0]=k
        i+=1
    print("we find the lowest (x1,x2) equals (",round(x1[0],2),",",round(x1[1],2),") after iterating 1000 times.")
gd_hw2_2([0,0],0.0001,1000)
gd_hw2_2([0,0],0.001,1000)
gd_hw2_2([0,0],0.01,1000)
gd_hw2_2([0,0],0.1,1000)
gd_hw2_2([0,0],1,1000)
gd_hw2_2([0,0],10,1000)
```

```python
"""Question 1d conclusion: From the iteration we ran, we can find when alpha equals 0.01 or 0.1, the x* we found is equal to the unique solution we found in question a. My conclusion is that when alpha is too small, we might not reach the local min before the iteration ends. When alpha is too big, the x* increase very fast and we will easily get large (x1,x2) and never find the min we need. Thus, in this case I think alpha= 0.1 or 0.01 is the best choice in this case."""
```
