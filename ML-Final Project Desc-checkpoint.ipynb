```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets
import pandas as pd
```

```python
a=pd.read_csv("all/test.csv")
b=pd.read_csv("all/gender_submission.csv")
x_train=pd.read_csv("all/train.csv")
x_test=pd.merge(a,b,on="PassengerId")
x_all=pd.concat([x_train,x_test])
```

```python
writer = pd.ExcelWriter('x_all.xlsx')
x_all.to_excel(writer,'Sheet1')
writer.save()
```

```python
ax = x_all["Sex"].value_counts().plot(kind='barh',figsize=(5,3))
ax.set_title("What is the distribution of Passenger Gender?", fontsize=18)
ax.set_xlabel("Number of Passengers", fontsize=15)
ax.set_xlim([0,1000])

# create a list to collect the plt.patches data
totals = []

# find the values and append to list
for i in ax.patches:
    totals.append(i.get_width())

# set individual bar lables using above list
total = sum(totals)

# set individual bar lables using above list
for i in ax.patches:
    # get_width pulls left or right; get_y pushes up or down
    ax.text(i.get_width()+.3, i.get_y()+.38, \
            str(round((i.get_width()/total)*100, 1))+'%', fontsize=14,
color='dimgrey')

# invert for largest on top 
ax.invert_yaxis()
```

```python
x_all["Fare"][x_all["Pclass"] == 1 ].plot(kind = 'hist', color='Green', normed=True)
x_all["Fare"][ x_all["Pclass"] == 2 ].plot(kind = 'hist',  color='Orange', normed=True)
x_all["Fare"][ x_all["Pclass"] == 3 ].plot(kind = 'hist',   color='Red', normed=True)
plt.title('Passenger Fare Based on Class Level')
plt.xlabel('Passenger Fare \n \n Green=First Class  Orange=Second Class  Red=Thrid Class')
```

```python
death_counts = pd.crosstab([x_all.Pclass, x_all.Sex], x_all.Survived.astype(bool))
death_counts.plot(kind='bar', stacked=True, color=['red','blue'])
plt.title('Percentage of survival rate in different gender and class combinations')
```
